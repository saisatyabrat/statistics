# Probability Distribution Function 

- it is a function that describe the probability of obtaining different values of random varriable in a perticulat probability distribution 
- it happens at discreat value

- types of PDF
- Probability Mass Function(PMF)
- Probability Density Function(PDF)

- it is a mathematical  function that describe the probability distribution of a discrete random varriable

- the probability assigned by the PMF must satisfy 2 condition
- (a)the probability should be possitive value or 0 it couldnot negetive value
- (b)the sum of probability assigned to all possible values must equal to 1

# Probability Mass Function(PMF)
- for single die roll

```python
pmf=[]
for i in range(1000):
    pmf.append(random.randint(1,6))
```

```python
s=(pd.Series(pmf).value_counts()/pd.Series(pmf).value_counts().sum()).sort_index()
```

```python
s
```

```python
s.plot(kind='bar')
plt.xlabel("Sample Space")
plt.ylabel("Probability")
plt.title("Probability mass function graph for a singe roll of a dice")
plt.show()
```

- for two die at a time

```python
mul=[]
for i in range(10000):
    a=random.randint(1,6)
    b=random.randint(1,6)
    mul.append(a+b)
```

```python
s2=(pd.Series(mul).value_counts()/pd.Series(pmf).value_counts().sum()).sort_index()
```

```python
s2.plot(kind='bar')
plt.xlabel("Sample Space")
plt.ylabel("Probability")
plt.title("Probability mass function graph for two times roll of a dice")
plt.show()
```

- for tossing  a coin 
- head=1, Tail=0

```python
coin=[]
for i in range(10000):
    coin.append(random.randint(0,1))
```

```python
s3=pd.Series(coin).value_counts()/pd.Series(coin).value_counts().sum()
```

```python
s3.plot(kind='bar')
plt.xlabel("Sample Space")
plt.ylabel("Probability")
plt.title("Probability mass function graph for tossing a coin H-0,T-1")
plt.show()
```

# Cumulative Distribution Function(CDF) of PMF

- for roll a dice

```python
cdf=[]
for i in range(10000):
    cdf.append(random.randint(1,6))
```

```python
b=pd.Series(cdf).value_counts().sort_index()
b
```

```python
pb=(pd.Series(cdf).value_counts()/pd.Series(cdf).value_counts().sum()).sort_index()
```

```python
final=np.cumsum(pb)
```

```python
final.plot(kind='bar')
plt.xlabel("Sample Space")
plt.ylabel("cumulative Probability")
plt.title("cumulative Distribution Function graph for single roll a dice")
plt.show()
```

- for rolling a dice twice

```python
sum1=[]
for i in range(10000):
    di1=random.randint(1,6)
    di2=random.randint(1,6)
    sum1.append(di1+di2)
    
```

```python
t=(pd.Series(sum1).value_counts()/pd.Series(sum1).value_counts().sum()).sort_index()
c=np.cumsum(t)
```

```python
c.plot(kind='bar')
plt.xlabel("Sample Space")
plt.ylabel("cumulative Probability")
plt.title("cumulative Distribution Function graph for roll two dice sum")
plt.show()
```

- for a coin

```python
coin=[]
for i in range(10000):
    coin.append(random.randint(0,1))
    
co=(pd.Series(coin).value_counts()/pd.Series(coin).value_counts().sum()).sort_index()
cum1=np.cumsum(co)
cum1.plot(kind='bar')
plt.xlabel("Sample Space")
plt.ylabel("cumulative Probability")
plt.title("cumulative Distribution Function graph for a coin")
plt.show()
```

# Probability Density Function

```python
from numpy.random import normal
```

```python
sample=normal(loc=50,scale=5,size=1000)
```

```python
sample.std()
```

```python
plt.hist(sample,bins=10)
```

```python
sample_mean=sample.mean()
sample_std=sample.std()
```

```python
print(sample_mean)
print(sample_std)
```

```python
from scipy.stats import norm
```

```python
dist=norm(sample_mean,sample_std)
```

```python
values=np.linspace(sample.min(),sample.max(),100)
```

```python
probability_density=[dist.pdf(value) for value in values]
```

```python
plt.hist(sample,bins=10,density=True)
plt.plot(values,probability_density)
```
